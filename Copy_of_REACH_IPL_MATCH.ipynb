```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("IPL_Matches_2008-2020.csv")

# Top 5 Winning Teams
top_winners = df['winner'].value_counts().head(5)

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=top_winners.values, y=top_winners.index, palette='viridis')
plt.title('Top 5 Teams with Most Wins (2008–2020)')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Most Played Venues
most_played_venues = df['venue'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=most_played_venues.values, y=most_played_venues.index, palette='magma')
plt.title('Top 10 Most Played Venues')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.tight_layout()
plt.show()
```

```python
# Toss Winner = Match Winner
toss_and_match_winner = (df['toss_winner'] == df['winner']).sum()
total_matches_with_winner = df['winner'].notna().sum()
toss_win_match_win_pct = round((toss_and_match_winner / total_matches_with_winner) * 100, 2)

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(
    [toss_and_match_winner, total_matches_with_winner - toss_and_match_winner],
    labels=['Toss Winner = Match Winner', 'Toss Winner ≠ Match Winner'],
    autopct='%1.1f%%',
    colors=['#4CAF50', '#F44336'],
    startangle=140
)
plt.title('Toss Winner vs Match Winner')
plt.axis('equal')
plt.show()

print(f"Percentage of matches where toss winner also won the match: {toss_win_match_win_pct}%")
```

```python
# Convert date and extract season
df['season'] = pd.to_datetime(df['date'], dayfirst=True).dt.year

# Matches per season
matches_per_season = df['season'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=matches_per_season.index, y=matches_per_season.values, marker='o')
plt.title('Matches Played per Season (2008–2020)')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Top 5 Toss Winning Teams
most_toss_winners = df['toss_winner'].value_counts().head(5)

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=most_toss_winners.values, y=most_toss_winners.index, palette='cool')
plt.title('Top 5 Teams with Most Toss Wins (2008–2020)')
plt.xlabel('Number of Toss Wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```
